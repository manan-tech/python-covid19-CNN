```python
from matplotlib import pyplot as plt
from tensorflow.keras.layers import Conv2D, Input, Dense, Flatten, Dropout
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Model
from tensorflow.keras.losses import sparse_categorical_crossentropy
import numpy as np
import itertools
from sklearn.metrics import confusion_matrix
from plot_cm import plot_confusion_matrix
%matplotlib inline
```

```python
(trainx, trainy), (testx, testy) = cifar10.load_data()
trainx, testx = trainx/255, testx/255
print(trainx.shape)
```

```python
# For Convolutions layer we need (n, h, w, color) the prev shape is okay for feed forward neural network
trainx = np.expand_dims(trainx, -1)
testx = np.expand_dims(testx, -1)
print(trainx.shape)
```

```python
K = len(set(trainy))
print(f"Number of classes: {K}")
```

```python
i = Input(shape=trainx[0].shape)
x = Conv2D(32, (3, 3), strides=2, activation='relu')(i)
x = Conv2D(64, (3, 3), strides=2, activation='relu')(x)
x = Conv2D(128, (3, 3), strides=2, activation='relu')(x)
x = Flatten()(x)
x = Dropout(0.5)(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(K, activation='softmax')(x)
model = Model(i, x)
```

```python
model.compile(optimizer='adam', loss=sparse_categorical_crossentropy, metrics = ['accuracy'])
r = model.fit(trainx, trainy, epochs=20, validation_data=(testx, testy))
```

```python
plt.plot(r.history['loss'], label='Loss')
plt.plot(r.history['val_loss'], label='Val_Loss')
plt.legend()
```

```python
plt.plot(r.history['accuracy'], label='Accuracy')
plt.plot(r.history['val_accuracy'], label='Val_Accuracy')
plt.legend()
```

```python
model.evaluate(testx, testy)
```

```python
ptest = model.predict(testx).argmax(axis=1)
cm = confusion_matrix(testy, ptest)
plot_confusion_matrix(cm, list(range(10)))
```

```python
misclassified_id = np.where(testy != ptest)[0]
i = np.random.choice(misclassified_id)
plt.imshow(testx[i], cmap='grey')
plt.title(f"True Label: {testy[i]}, Predicted: {ptest[i]}")
```
